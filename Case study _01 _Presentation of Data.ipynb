### 3.Plot a bar graph of male vs female buyers participated in the sales 

```python
# Counting the number of male and female participants
```

```python
print(data['Buyer Gender'].tolist().count('Male'))
```

```python
print(data['Buyer Gender'].tolist().count('Female'))
```

```python
# plotting the bar graph of male vs female buyers participated in the sales
```

```python
series = data["Buyer Gender"].value_counts()

series.plot(kind='bar')
plt.title('Male and Female buyers participated in the sales')
plt.xlabel('Buyer Gender')
plt.ylabel('Number of participants')
plt.show()
```

The graph shows the number of Male and Female buyers participated in the sales.

### 4.Find the top 5 cars based on their sales price. 

```python
#Sorting the sale price from highest to lowest to find the top 5 cars with higher sales price
```

```python
data.sort_values('Sale Price',ascending=False).head()
```
